```python
model_df = df[df["valid_label_h1"]].copy()

model_df = model_df.sort_values(
    ["retailer", "product_id", "date"]
).reset_index(drop=True)

model_df.shape
```

```python
model_df["dayofweek"] = model_df["date"].dt.dayofweek
model_df["weekofyear"] = model_df["date"].dt.isocalendar().week.astype("int16")
model_df["month"] = model_df["date"].dt.month
model_df["day"] = model_df["date"].dt.day

model_df[["date", "dayofweek", "weekofyear", "month", "day"]].head()
```

```python
group_cols = ["retailer", "product_id"]

model_df["price_lag_1"] = (
    model_df
    .groupby(group_cols, observed=True)["price"]
    .shift(1)
)

model_df["price_lag_2"] = (
    model_df
    .groupby(group_cols, observed=True)["price"]
    .shift(2)
)

model_df["price_lag_3"] = (
    model_df
    .groupby(group_cols, observed=True)["price"]
    .shift(3)
)

model_df["price_change_pct_lag_1"] = (
    ((model_df["price"] - model_df["price_lag_1"]) / model_df["price_lag_1"]) * 100
)

model_df["price_change_pct_lag_2"] = (
    ((model_df["price_lag_1"] - model_df["price_lag_2"]) / model_df["price_lag_2"]) * 100
)

model_df[[
    "retailer",
    "product_id",
    "date",
    "price",
    "price_lag_1",
    "price_lag_2",
    "price_lag_3",
    "price_change_pct_lag_1",
    "price_change_pct_lag_2",
    "target_direction_h1_t05"
]].head(20)
```

```python
model_df
```

```python
gap_mask = model_df["capture_gap_before_row"].fillna(False)

model_df.loc[gap_mask, [
    "price_lag_1",
    "price_lag_2",
    "price_lag_3",
    "price_change_pct_lag_1",
    "price_change_pct_lag_2"
]] = np.nan

model_df[[
    "retailer",
    "product_id",
    "date",
    "price",
    "capture_gap_before_row",
    "price_lag_1",
    "price_lag_2",
    "price_lag_3",
    "price_change_pct_lag_1",
    "price_change_pct_lag_2"
]].head(20)
```

```python
group_cols = ["retailer", "product_id"]

model_df["rolling_price_mean_3"] = (
    model_df
    .groupby(group_cols, observed=True)["price"]
    .transform(lambda x: x.rolling(window=3, min_periods=2).mean())
)

model_df["rolling_price_std_3"] = (
    model_df
    .groupby(group_cols, observed=True)["price"]
    .transform(lambda x: x.rolling(window=3, min_periods=2).std())
)

model_df["rolling_price_mean_7"] = (
    model_df
    .groupby(group_cols, observed=True)["price"]
    .transform(lambda x: x.rolling(window=7, min_periods=3).mean())
)

model_df["rolling_price_std_7"] = (
    model_df
    .groupby(group_cols, observed=True)["price"]
    .transform(lambda x: x.rolling(window=7, min_periods=3).std())
)

model_df[[
    "price",
    "rolling_price_mean_3",
    "rolling_price_std_3",
    "rolling_price_mean_7",
    "rolling_price_std_7"
]].head(20)
```

```python
model_df["price_vs_rolling_mean_3_pct"] = (
    ((model_df["price"] - model_df["rolling_price_mean_3"]) / model_df["rolling_price_mean_3"]) * 100
)

model_df["price_vs_rolling_mean_7_pct"] = (
    ((model_df["price"] - model_df["rolling_price_mean_7"]) / model_df["rolling_price_mean_7"]) * 100
)

model_df[[
    "price",
    "rolling_price_mean_3",
    "price_vs_rolling_mean_3_pct",
    "rolling_price_mean_7",
    "price_vs_rolling_mean_7_pct"
]].head(20)
```

```python
model_df["price_changed_lag_1"] = (
    model_df["price_change_pct_lag_1"].fillna(0).abs() > 0
).astype("int8")

model_df["price_changed_lag_2"] = (
    model_df["price_change_pct_lag_2"].fillna(0).abs() > 0
).astype("int8")

model_df["recent_change_count_3"] = (
    model_df
    .groupby(group_cols, observed=True)["price_changed_lag_1"]
    .transform(lambda x: x.rolling(window=3, min_periods=1).sum())
)

model_df["recent_change_count_7"] = (
    model_df
    .groupby(group_cols, observed=True)["price_changed_lag_1"]
    .transform(lambda x: x.rolling(window=7, min_periods=1).sum())
)

model_df[[
    "price_change_pct_lag_1",
    "price_changed_lag_1",
    "price_changed_lag_2",
    "recent_change_count_3",
    "recent_change_count_7"
]].head(20)
```

```python
group_cols = ["retailer", "product_id"]

model_df["rolling_price_mean_3"] = (
    model_df
    .groupby(group_cols, observed=True)["price"]
    .transform(lambda x: x.shift(1).rolling(window=3, min_periods=2).mean())
)

model_df["rolling_price_std_3"] = (
    model_df
    .groupby(group_cols, observed=True)["price"]
    .transform(lambda x: x.shift(1).rolling(window=3, min_periods=2).std())
)

model_df["rolling_price_mean_7"] = (
    model_df
    .groupby(group_cols, observed=True)["price"]
    .transform(lambda x: x.shift(1).rolling(window=7, min_periods=3).mean())
)

model_df["rolling_price_std_7"] = (
    model_df
    .groupby(group_cols, observed=True)["price"]
    .transform(lambda x: x.shift(1).rolling(window=7, min_periods=3).std())
)

model_df[[
    "price",
    "price_lag_1",
    "rolling_price_mean_3",
    "rolling_price_std_3",
    "rolling_price_mean_7",
    "rolling_price_std_7"
]].head(20)
```

```python
model_df["price_vs_rolling_mean_3_pct"] = (
    ((model_df["price"] - model_df["rolling_price_mean_3"]) / model_df["rolling_price_mean_3"]) * 100
)

model_df["price_vs_rolling_mean_7_pct"] = (
    ((model_df["price"] - model_df["rolling_price_mean_7"]) / model_df["rolling_price_mean_7"]) * 100
)

model_df[[
    "price",
    "rolling_price_mean_3",
    "price_vs_rolling_mean_3_pct",
    "rolling_price_mean_7",
    "price_vs_rolling_mean_7_pct"
]].head(20)
```

```python
model_df["change_event"] = model_df["price_changed_lag_1"] == 1

model_df["last_change_date"] = model_df["date"].where(model_df["change_event"])

model_df["last_change_date"] = (
    model_df
    .groupby(group_cols, observed=True)["last_change_date"]
    .ffill()
)

model_df["days_since_last_change"] = (
    model_df["date"] - model_df["last_change_date"]
).dt.days

model_df["days_since_last_change"] = model_df["days_since_last_change"].fillna(9999)

model_df[[
    "retailer",
    "product_id",
    "date",
    "price",
    "price_changed_lag_1",
    "last_change_date",
    "days_since_last_change"
]].head(30)
```

```python
feature_cols = [
    # current known price state
    "price",
    "days_since_previous_capture",

    # lag features
    "price_lag_1",
    "price_lag_2",
    "price_lag_3",
    "price_change_pct_lag_1",
    "price_change_pct_lag_2",

    # rolling features
    "rolling_price_mean_3",
    "rolling_price_std_3",
    "rolling_price_mean_7",
    "rolling_price_std_7",
    "price_vs_rolling_mean_3_pct",
    "price_vs_rolling_mean_7_pct",

    # recent change-history features
    "price_changed_lag_1",
    "price_changed_lag_2",
    "recent_change_count_3",
    "recent_change_count_7",
    "days_since_last_change",

    # calendar features
    "dayofweek",
    "weekofyear",
    "month",
    "day",

    # categorical/static features
    "retailer",
    "category",
    "own_brand_flag",
]

target_col = "target_direction_h1_t05"

model_features_df = model_df[
    ["product_id", "date"] + feature_cols + [target_col]
].copy()

model_features_df.shape
```

```python
missing_summary = (
    model_features_df
    .isna()
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

missing_summary.columns = ["column", "missing_count"]
missing_summary["missing_share"] = missing_summary["missing_count"] / len(model_features_df)

missing_summary
```
